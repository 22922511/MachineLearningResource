```python
import numpy as np
import pandas as pd
import datetime
```

```python
df=pd.read_csv("messages.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.info()
```

#Extract year

```python

df["year"]=df["date"].dt.year
```

```python
df.head()
```

#Extract Month

```python
df["month"]=df["date"].dt.month
```

```python
df.head()
```

#Extract month name

```python
df["month_name"]=df["date"].dt.month_name()
```

```python
df.head()
```

#Extract day

```python
df["day"]=df["date"].dt.day
df.head()
```

#Extract day of week

```python
df["dayofweeek"]=df["date"].dt.dayofweek
df.head()
```

#Extract name of day

```python
df["name ofday"]=df["date"].dt.day_name()
df.head()
```

```python
df["weekend"]=np.where(df["dayofweeek"].isin([6,5]),1,0)
df.head()
```

#Extract quarter

```python
df["quarter"]=df["date"].dt.quarter
df.head()
```

#Current date time

```python
print(datetime.datetime.today())
```

```python
df["hour"]=df["date"].dt.hour
df["minutes"]=df["date"].dt.minute
df["seconds"]=df["date"].dt.second
df.sample(5)
```
